This iPython Notebook contains code and figures quantitatively demonstrating critical slowing down, and some associated early warning signals (autocorrelation, variance and skew) in a dynamical system approaching a fold bifurcation. It can be viewed in the browser, or caeasily downloaded and executed by the reader.

The equation used is the fold bifurcation $\dot{x} = -x^3 + 3x + r$ (where $r$ is the control parameter) with critical points at $r = ±2$. For two points (far from and approaching the bottom bifurcation point) 

Firstly, the dynamical system portrait ($\dot{x} = f(x)$) and fixed point basin of attraction potentials ($V(x) = -\int{f(x) dx}$) are sketched. 

Further, various time series are produced, including a slowly increasing $r$ and stochastic noise $\eta$. These are not aimed to be a rigorous analysis, but to give intuition behind critical slowing, how this affects early warning signals, and how this is critically dependent on the assumption of distinct process timescales (bifurcation parameter drift, decay to equilibrium and stochastic noise). These timescales are given quantified order-of-magnitude estimates throughout.

N.B. The figures shown in this notebook are likely to differ slightly from those included in the poster/presentation, owing to the stochastic nature of the simulations.

### General helper functions to generate ODEs for our fold model parameterisation

```python
def fold_bifurcation(r):
    ode = lambda x: -x**3 + 3*x + r
    potential = lambda x: 0.25*(x**4) - 1.5*(x**2) - r*x
    
    return ode,potential
```

```python
def show_ode(r, x_pos=None, xlim=(-5,5), ylim=(-12,10)):
    x = np.linspace(xlim[0], xlim[1], 100)
    plt.axhline(0, color='black')
    ode,potential = fold_bifurcation(r)
    plt.plot(x, ode(x))
    plt.xlim(xlim)
    plt.ylim(ylim)
    
    if x_pos:
        plt.axvline(x_pos, color='green', linestyle=':')
```

```python
show_ode(-10, x_pos=1)
```

```python
def show_potential_well(r, x_pos=None, xlim=(-5,5), ylim=(-25,25)):
    x = np.linspace(xlim[0], xlim[1], 100)
    ode,potential = fold_bifurcation(r)
    plt.plot(x, potential(x))
    plt.xlim(xlim)
    plt.ylim(ylim)
    
    if x_pos:
        y_pos = potential(x_pos)
        plt.scatter([x_pos], [y_pos], s=[75], marker='o', color='green')
```

## Stochastic Simulations

Helpers functions

```python
def step(ode, x, time_scaling=0.025):
    '''
    Step forward one time step using Euler's method: 
    time_scaling : implicit conversion between natural units of x and time steps 
    '''
    h = time_scaling * ode(x)
    return x + h
```

```python
def simulation(ode, x_0, noise, num_runs, p_noise=1./25):
    '''
    Generate a time series from ode, with no drifting parameter.
    
    If noise is set, then a Gaussian additive perturbance of N(0,noise)
    is added with uniform probability p_noise.
    '''
    values = [x_0]
    
    current_x = x_0
    for i in range(num_runs):
        new_x = step(ode, current_x)
        if noise > 0 and np.random.rand() < p_noise:
            new_x += np.random.normal(0,noise)
        values.append(new_x)
        current_x = new_x
    
    return values
```

```python
def simulation_drift(x_0, n, noise=0.01, p_noise=1./25):
    '''
    Generate a time series from ode, with a drifting parameter.
    '''
    
    values = np.zeros((n,1))
    
    r_min = -10
    r_max = 1.5
    
    r_values = np.linspace(r_min, r_max, n)
    
    current_x = x_0 
    for i,r in enumerate(r_values):
        ode,potential = fold_bifurcation(r)
        new_x = step(ode, current_x)
        if noise > 0 and np.random.rand() < p_noise:
            new_x += np.random.normal(0,noise)
        values[i,0] = new_x
        current_x = new_x
    
    return (r_values, [x[0] for x in values])
```

```python
def show_acorr(x):
    '''
    Show a scatterplot with linear fit and Pearson correlation coefficient
    between X_t and X_{t-1} (i.e. lag-1 autocorrelation) for a time series.
    '''
    
    n = len(x)
    y = x[:n-1]
    z = x[1:n]
    
    fit = np.polyfit(y, z, 1)
    fit_fn = poly1d(fit)
    
    correlation = np.corrcoef(y,z)[1,0]
    
    line_y = np.linspace(np.min(y),np.max(y),2)
    
    plt.plot(line_y, fit_fn(line_y), 'r', label='r=%0.3f' % correlation)
    
    plt.scatter(y,z, alpha=0.5)
    plt.xlim(np.min(y), np.max(y))
    plt.ylim(np.min(z), np.max(z))
    plt.legend(loc='lower right')
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab

def show_distribution(x, xlim=0.1):
    '''
    Show a histogram of values from a time series, and a best-fit Gaussian 
    p.d.f. to qualitatively demonstrate variance and skewness properties.
    '''
    lower = np.mean(x) - xlim
    upper = np.mean(x) + xlim
    
    filtered = filter(lambda x: x > lower and x < upper, x)
    n, bins, patches = plt.hist(filtered, bins=25, alpha=0.5)
    
    (mu, sigma) = norm.fit(filtered)
    y = mlab.normpdf( bins, mu, sigma)
    l = plt.plot(bins, y, 'r--', linewidth=1)
    
    plt.xlim(lower, upper)
```

## Simple Simulations

### Straightforward linear decay to a stable point -- i.e. no stochastic forcing
Decays back to fixed point in $\tau_\lambda \approx 5$ steps (with time scaling 0.025), when we are far away from the basin. Although the decay becomes closer to $\tau_\lambda \approx 40$ as we get closer to the bifurcation.

Note: the actual process of decay is what has high autocorrelation and skewness, so if we just take these first 5 steps, then we do see high autocorrelation and skewness! The reason these change as we reach an incipient bifurcation point is because more of our time is spent decaying (as opposed to being static in the well of the basin)

### Near bifurcation point
$r=1.5$: the time series plot shows slow return to equilibrium, high autocorrelation ($\approx 0.9$) and high variance (although variance/skew seem less convincing when the bifurcation parameter isn't drifting!)

Note that the stochastic forcing operates on a slower timescale than the decay -- we have at every time step $P(perturbance) = \frac{1}{25}$, so we expect a perturbance every 25 time steps, i.e. $\tau_\eta \approx 25 \gg \tau_\lambda \approx 5$.
